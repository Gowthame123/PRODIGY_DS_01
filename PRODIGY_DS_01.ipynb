```python
# Fill NaN values only in numeric columns
men = men.fillna(men.select_dtypes(include=['number']).mean())

# Convert only float columns to integer
men = men.astype({col: 'int' for col in men.select_dtypes('float').columns})
men.head()
```

```python
women = pd.read_csv(r"C:\Users\gowth\Downloads\PRODIGY_internship\dataset\task 1\API_SP.POP.TOTL.FE.IN_DS2_en_CSV_v2_7318\women_population.csv")
```

```python
women.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis = 1 , inplace = True)
```

```python
women.columns
```

```python
women.info()
```

```python
women = women.fillna(women.select_dtypes(include=['number']).mean())

# Convert only float columns to integer
women = women.astype({col: 'int' for col in women.select_dtypes('float').columns})
women.head()
```

```python
year_sums = {}
for column in men.columns:
   year_sums[column] = men[column].sum()

# Convert the result into a DataFrame for visualization
df_sums = pd.DataFrame(list(year_sums.items()), columns=['Year', 'Sum'])

df_sums.head()
```

```python


# Plotting the line chart using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_sums, x='Year', y='Sum', marker='o', color='skyblue')

# Adding title and labels
plt.title('Sum of men Population by Year')
plt.xlabel('Year')
plt.ylabel('Sum of Population')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_sums, x='Year', y='Sum', palette='Blues_d')
plt.title('Sum of Men Population by Year')
plt.xlabel('Year')
plt.ylabel('Sum of Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
